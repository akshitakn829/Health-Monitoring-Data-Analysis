```python
import pandas as pd
import matplotlib.pyplot as plt

# Health Monitoring Dataset
data = {
    "Name": ["Asha","Rahul","Priya","Kiran","Arun"],
    "Age": [22,35,28,45,50],
    "HeartRate": [72,88,76,95,102],
    "SystolicBP": [110,130,120,145,150],
    "DiastolicBP": [70,85,80,95,98],
    "BMI": [21.5,27.8,23.0,30.2,32.1]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Average values
print("\nAverage Heart Rate:", df["HeartRate"].mean())
print("Average BMI:", df["BMI"].mean())

# BMI Classification
def bmi_status(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["HealthStatus"] = df["BMI"].apply(bmi_status)

print("\nHealth Status:")
print(df[["Name","HealthStatus"]])

# Plot graph
plt.plot(df["Name"], df["HeartRate"])
plt.title("Heart Rate Monitoring")
plt.xlabel("Patient")
plt.ylabel("Heart Rate")
plt.show()
```
